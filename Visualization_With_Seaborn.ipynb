# Importing libraries

```python
#importing libraries
import seaborn as sns
sns.set()
sns.set(style='darkgrid')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Inserting data frame

```python
big_mart_data=pd.read_csv('/content/gdrive/MyDrive/machine learning /analytics_vidhya /bigmart_data.csv')
```

```python
big_mart_data.dropna(how='any',inplace=True)
big_mart_data.head()
```

# Creating a basic plots
### 1. Line chart

```python
sns.lineplot(x='Item_Weight',y='Item_MRP',data=big_mart_data);
```

```python
sns.lineplot(x='Item_Weight',y='Item_MRP',data=big_mart_data[:100]);
```

### Bar chart

```python
plt.figure(figsize=(30,5))
sns.barplot(x='Item_Type',y='Item_MRP',data=big_mart_data);
```

### Histogram

```python
sns.distplot(big_mart_data[['Item_MRP']]);
```

### Box plot

```python
sns.boxplot(data=big_mart_data['Item_Outlet_Sales']);
```

### Violin plot

```python
sns.violinplot(data=big_mart_data['Item_Outlet_Sales'],orient='verticle',color='lightblue')
```

### Scatter plots

```python
sns.relplot(x='Item_MRP',y='Item_Outlet_Sales',data=big_mart_data[:500]);
```

### Hue semantic

```python
sns.relplot(x='Item_MRP',y='Item_Outlet_Sales',hue='Item_Type', data=big_mart_data[:500])
```

```python
sns.lineplot(x='Item_Weight',y='Item_MRP',hue='Outlet_Size',data=big_mart_data[:150]);
```

### Bubble plot

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=big_mart_data[:200], kind="scatter", size="Item_Visibility", hue="Item_Visibility");
```

```python
sns.relplot(x='Item_Weight',y='Item_Visibility',hue='Outlet_Size',size='Outlet_Size',col='Outlet_Size',data=big_mart_data[:100])
```
